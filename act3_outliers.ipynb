Sustituir las primeras 5 columnas con m√©todo de desviaci√≥n est√°ndar

```python
#Outliers en boxplot=1
#Encontrar outliers
outliers_y1=data2[(y1>ls1)|(y1<li1)]
outliers_y1
```

```python
#df con y1 sin outliers
data_clean_y1=data2[(y1<=ls1)&(y1>=li1)] #Eliminar esa fila
data_clean_y1
```
